```python
from copy import copy, deepcopy

class Cash():
    def __init__(self):
        pass
    
class CLP(Cash):
    def __init__(self,
        n=None,
        ):
        self.n = n
    
    def __repr__(self):
        return f'{self.n}CLP'

    def __add__(self, other):
        if self==0:
            return CLP(other.n)
        if other==0:
            return CLP(self.n)
        return CLP(self.n+other.n)

    def __radd__(self, other):
        return self+other
    
    def __sub__(self, other):
        return CLP(self.n-other.n)
    
    def __copy__(self):
        return CLP(self.n)
    
class Wallet():
    def __init__(self, prices_dict):
        self.prices_dict = deepcopy(prices_dict)
        self.reset()
        
    def reset(self):
        self.bought_objs = []
        self.sold_objs = []
        
    def buy(self, obj, date):
        self.bought_objs += [obj]
        
    def sell(self, obj, date):
        self.sold_objs += [obj]
    
    def get_bought_final_prices(self):
        final_prices = [obj.final_price for obj in self.bought_objs]
        if len(final_prices)>0:
            return sum(final_prices)
        else:
            return CLP(0)
    
    def get_sold_final_prices(self):
        final_prices = [obj.final_price for obj in self.sold_objs]
        if len(final_prices)>0:
            return sum(final_prices)
        else:
            return CLP(0)
    
    def __repr__(self):
        print(self.bought_objs)
        print(self.sold_objs)
        txt = ''
        b = self.get_bought_final_prices()
        txt += f'bought={b}'+'\n'
        s = self.get_sold_final_prices()
        txt += f'sold={s}'+'\n'
        txt += f'total={s-b}'
        txt += f'---'+'\n'
        return txt
    
    def del_objs(self, to_del_objs):
        for obj in to_del_objs:
            idx = self.bought_objs.index(obj)
            self.bought_objs.pop(idx)
        
    def _sell(self, obj):
        to_del_objs = [obj]
        self.sold_objs += [obj]
        self.del_objs(to_del_objs)
        
    def sell(self, name):
        to_del_objs = []
        for obj in self.bought_objs:
            if name==obj.name:
                self.sold_objs += [obj]
                to_del_objs += [obj]
        self.del_objs(to_del_objs)
            
    def sell_all(self):
        to_del_objs = []
        for obj in self.bought_objs:
            sold_obj = copy(obj)
            
            new_price = CLP(self.prices_dict[obj.name].n*obj.n)
            sold_obj.set_price(new_price)
            
            self.sold_objs += [sold_obj]
            to_del_objs += [obj]
        #self.del_objs(to_del_objs)
        pass
    
class MarketObject():
    def __init__(self, name, n, final_price):
        self.name = name
        self.n = n
        self.final_price = final_price
        self.reset()
        
    def reset(self):
        pass
    
    def set_price(self, final_price):
        self.final_price = final_price
    
    def __copy__(self):
        return MarketObject(self.name, self.n, copy(self.final_price))
    
    def __repr__(self):
        txt = f'{self.n}~{self.name}({self.final_price})'
        return txt

prices = {
    'LTM':CLP(1950),
    'ENJOY':CLP(4.466),
    'ENELCHILE':CLP(42.93),
    'CAP':CLP(13183.86),
}
wallet = Wallet(prices)
wallet.buy(MarketObject('LTM', 400, CLP(507707)), '')
#wallet.buy(MarketObject('ENJOY', 42000, CLP(505406)), '')
#wallet.buy(MarketObject('ENJOY', 25000, CLP(310823)), '')
#wallet.buy(MarketObject('ENJOY', 20000, CLP(152080)), '')
wallet.buy(MarketObject('ENELCHILE', 2000, CLP(93023)), '5/20/201')
#wallet.buy(MarketObject('ENJOY', 42000, CLP(505406)), '')
wallet.buy(MarketObject('ENELCHILE', 6988, CLP(304224)), '5/29/2021')
wallet.buy(MarketObject('CAP', 23, CLP(303229)), '5/29/2021')

print(wallet)
wallet.sell_all()
#wallet.sell('CAP')
print(wallet)
#wallet.get_final_prices()
```
